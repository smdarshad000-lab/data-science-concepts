```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./rock_density_xray.csv')
```

```python
df.head()
```

```python
df.columns = ['Signal', 'Density']
```

```python
sns.scatterplot(x='Signal', y='Density', data=df)
```

```python
X = df['Signal'].values.reshape(-1, 1)
y = df['Density']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_preds = lr_model.predict(X_test)
```

```python
lr_preds
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, lr_preds) #mean absolute value
```

```python
np.sqrt(mean_squared_error(y_test, lr_preds)) #root mean squared error
```

```python
signal_range = np.arange(0, 100)
```

```python
signal_range
```

```python
signal_preds = lr_model.predict(signal_range.reshape(-1, 1))
```

```python
signal_preds
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(x='Signal', y='Density', data=df)

plt.plot(signal_range, signal_preds) #this model have a problem
# as it is a linear model it will not be able to capture the non linear relationship
```

```python

def run_model(model, X_train, y_train, X_test, y_test):
    #FIT model training
    model.fit(X_train, y_train)
    #Make predictions
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    print(f'RMSE: {rmse}, MAE: {mae}')
    signal_range = np.arange(0, 100)
    signal_preds = model.predict(signal_range.reshape(-1, 1))

    plt.figure(figsize=(12,8), dpi=200)
    sns.scatterplot(x='Signal', y='Density', data=df, color='green')
    plt.plot(signal_range, signal_preds)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=5), LinearRegression())
# degree of 6 is satisfying in this case
```

```python
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k_values = [1,5,10,20,30]


for n in k_values: 

    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svr = SVR()
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': ['auto', 'scale'],
    'kernel': ['rbf', 'linear']
}

grid = GridSearchCV(svr, param_grid)

run_model(grid, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor  
```

```python
rfr = RandomForestRegressor(n_estimators=10)
```

```python
run_model(rfr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
```

```python
model = GradientBoostingRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```

```python
model = AdaBoostRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```
