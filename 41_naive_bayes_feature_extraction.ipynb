```python
import pandas as pd
```

```python
with open('One.txt') as mytext:
    words_one = mytext.read().lower().split()
    uni_words_one = set(words_one)
```

```python
uni_words_one
```

```python
with open('Two.txt') as mytext:
    words_two = mytext.read().lower().split()
    uni_words_two = set(words_two)
```

```python
uni_words_two
```

```python
all_uni_words = set()
all_uni_words.update(uni_words_one)
```

```python
all_uni_words.update(uni_words_two)
```

```python
all_uni_words #now this holds all the unique words in both files
```

```python
full_vocab = dict()
i=0

for word in all_uni_words:
    full_vocab[word] = i
    i+=1
print(full_vocab) # this is our vocabulary
```

```python
one_freq = [0]*len(full_vocab)
two_freq = [0]*len(full_vocab)
all_words = ['']*len(full_vocab)
```

```python
one_freq
```

```python
with open('One.txt') as mytext:
    one_text = mytext.read().lower().split()
```

```python
for word in one_text:
    word_ind = full_vocab[word]
    one_freq[word_ind] += 1
```

```python
one_freq
```

```python
with open('Two.txt') as mytext:
    two_text = mytext.read().lower().split()
```

```python
two_text
```

```python
for word in two_text:
    word_ind = full_vocab[word]
    two_freq[word_ind] += 1
```

```python
two_freq
```

```python
all_words
```

```python
for word in full_vocab:
    word_ind = full_vocab[word]
    all_words[word_ind] = word
```

```python
all_words
```

```python
bow = pd.DataFrame([one_freq, two_freq], columns = all_words) #creating a dataframe using pandas
```
