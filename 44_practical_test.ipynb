```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./KNNAlgorithmDataset.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(20,12))
# Use only numeric columns for a heatmap (e.g., correlation matrix)
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['Unnamed: 32'], errors='ignore')
corr = numeric_df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation heatmap')
plt.show()
```
