```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('advertising.csv')
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
Y = df['sales']
```

```python
# Y = mx+b
# Y = B1X + B0
np.polyfit(X, Y, deg=1)
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
# Y = MX+B
# here it is : B1 * potential_spend + B0
predicted_sale = 0.04868788 * potential_spend + 4.24302822
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(potential_spend, predicted_sale, color='red')
```

```python
# Y = B1X+ B0
# so now  we have multiple values so we have to build a multiple linear regression model so the equation will be:
# Y = B3x ** 3 + B2x ** 2 + B1x + B0
```

```python
pot_spend = np.linspace(0, 500, 100)
```

```python
pred_sales =  3.07615033e-07 * pot_spend **3 + -1.89392449e-04 * pot_spend ** 2 + 8.20886302e-02 * pot_spend +  2.70495053e+00
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(pot_spend, pred_sales, color='red')
```
