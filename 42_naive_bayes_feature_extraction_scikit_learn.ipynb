```python
text = ['This is a line', 
        'This is another line',
        'Completely different line']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
#cv = CountVectorizer(stop_words='english') #the stop_words are the words we want to remove, basically do not
#hold much importance in the sentence analysis

cv = CountVectorizer()
```

```python
sparse_matrix = cv.fit_transform(text) # sparse matrix
```

```python
cv.vocabulary_
```

```python
sparse_matrix.todense()
```

```python
tfidf = TfidfTransformer() # term frequency inverse document frequency helps in normalizing the data
```

```python
sparse_matrix
```

```python
results = tfidf.fit_transform(sparse_matrix) # sparse matrix of tfidf
```

```python
results.todense()
```
