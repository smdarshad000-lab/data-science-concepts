```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
X = df.drop('sales',axis=1)
```

```python
df
```

```python
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
Polynomial_convertor = PolynomialFeatures(degree=3, include_bias=False)
```

```python
X
```

```python
poly_features = Polynomial_convertor.fit_transform(X)
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( poly_features, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(X_train)
```

```python
X_train = scalar.transform(X_train) #scaling down to very less cofficience
```

```python
X_test = scalar.transform(X_test)
```

```python
X_train
```

# Implementing the rigde model (L2 regression)

```python
from sklearn.linear_model import Ridge
```

```python
help(Ridge)
```

```python
ridge_model = Ridge(alpha=10) #honestly we do not know that alpha is equal to 10
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
RMSE
```

```python
from sklearn.linear_model import RidgeCV
```

```python
# 1. import RidgeCV
# 2. Choose your alphas, and choose your scoring,
# 3.  Let it do its work. 

ridge_CV_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')
```

```python
ridge_CV_model.fit(X_train, y_train)
```

```python
ridge_CV_model.alpha_
```

```python
from sklearn.metrics import get_scorer_names
```

```python
get_scorer_names()
```

```python
test_predictions = ridge_CV_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
MAE
```

```python
RMSE #Here by using the Rigde regression, we found out better RMSE and MAE than before. 
```

```python
ridge_CV_model.best_score_
```

# LASSO - Least absolute shrinkage and selection operator (L1 regression)

```python
from sklearn.linear_model import LassoCV
```

```python
LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.1, n_alphas=100, cv=5) #here we created the model
```

```python
lasso_cv_model.fit(X_train, y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
RMSE =np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
MAE
```

```python
lasso_cv_model.coef_ #lasso is considering only two features 
```
